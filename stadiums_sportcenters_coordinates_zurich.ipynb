```python
import requests
import json
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from urllib.request import urlopen 
```

```python
url = "http://overpass-api.de/api/interpreter?data=%5Bout%3Ajson%5D%3B%0A%28%0A%09area%283601682248%29-%3E.searchArea%3B%0A%20%20way%5B%22leisure%22~%22stadium%7Csports_centre%22%5D%28area.searchArea%29%3B%0A%0A%20%20%0A%29%3B%0A%2F%2F%20print%20results%0Aout%20center%3B%0A%0A%0A"
#[out:json];
#(
#	{{geocodeArea:Zurich}}->.searchArea;
#  way["leisure"~"stadium|sports_centre"](area.searchArea);  
#);
#// print results
#out center;
```

```python
json_url = urlopen(url)
data = json.loads(json_url.read())
```

```python
data['elements'][1]
```

```python
list(data['elements'][1].keys())
```

```python
# Collect coords into list
coords = []
for element in data['elements']:
    if 'center' in list(element.keys()):
        if 'lon' in list(element['center'].keys()):
            lon = element['center']['lon']
        else:
            lon = ""
        if 'lat' in list(element['center'].keys()):
            lat = element['center']['lat']
        else:
            lat = ""
    else:
        lon = ""
        lat = ""
    if 'addr:city' in list(element['tags'].keys()):
        addr_city = element['tags']['addr:city']
    else:
        addr_city = ""
    if 'leisure' in list(element['tags'].keys()):
        leisure = element['tags']['leisure']
    else:
        leisure = ""
    if 'name' in list(element['tags'].keys()):
        name = element['tags']['name']
    else:
        name = "no"
    if 'sport' in list(element['tags'].keys()):
        sport = element['tags']['sport']
    else:
        sport=""

    coords.append((name, addr_city, lon, lat, leisure, sport))
 
```

```python
coords
```

```python
df = pd.DataFrame(coords, columns=("name", "addr_city", "lon", "lat", "leisure", "sport"))
```

```python
df.info()
```

```python
df['coordinates'] = "(" + df['lon'].map(str) + " , "  + df['lat'].map(str)+")" 
```
