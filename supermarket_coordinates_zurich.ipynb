```python
# Collect coords into list
coords = []
for element in data['elements']:
  if element['type'] == 'node':
    lon = element['lon']
    lat = element['lat']
    if 'name' in list(element['tags'].keys()):
        name = element['tags']['name']
    else:
        name=" "
    if 'addr:city' in list(element['tags'].keys()):
        city = element['tags']['addr:city']
    else:
        city=" "
    if 'addr:postcode' in list(element['tags'].keys()):
        postcode = element['tags']['addr:postcode']
    else:
        postcode=" "
    if 'addr:street' in list(element['tags'].keys()):
        street = element['tags']['addr:street']
    else:
        street=" "

    coords.append((name, lon, lat, city, postcode, street))
  #elif 'center' in element:
  #lon = element['center']['lon']
  #lat = element['center']['lat']  
  #nm = element['tags']['name']
  #coords.append((lon, lat))
 
```

```python
df = pd.DataFrame(coords, columns=("name", "lon", "lat", "city", "postcode", "street"))
```
