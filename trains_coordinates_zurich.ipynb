```python
import requests
import json
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from urllib.request import urlopen 
```

```python
url = "http://overpass-api.de/api/interpreter?data=%5Bout%3Ajson%5D%3B%0A%28%0A%09area%283601682248%29-%3E.searchArea%3B%0A%20%20node%5B%22railway%22%3D%22station%22%5D%28area.searchArea%29%3B%0A%0A%20%20%0A%29%3B%0A%2F%2F%20print%20results%0Aout%20center%3B"

#[out:json];
#(
#	{{geocodeArea:Zurich}}->.searchArea;
#  node["railway"="station"](area.searchArea);  
#);
#out center;
```

```python
json_url = urlopen(url)
data = json.loads(json_url.read())
```

```python
data['elements'][1]
```

```python
# Collect coords into list
coords = []
for element in data['elements']:
  if element['type'] == 'node':
    lon = element['lon']
    lat = element['lat']
    if 'name' in list(element['tags'].keys()):
        name = element['tags']['name']
    else:
        name = "no"
    if 'funicular' in list(element['tags'].keys()):
        funicular = element['tags']['funicular']
    else:
        funicular = "no"
    if 'train' in list(element['tags'].keys()):
        train = element['tags']['train']
    else:
        train = "no"
    if 'network' in list(element['tags'].keys()):
        network = element['tags']['network']
    else:
        network=" "
    if 'operator' in list(element['tags'].keys()):
        operator = element['tags']['operator']
    else:
        operator=" "

    coords.append((name, funicular, lon, lat, network, operator))
 
```

```python
coords
```

```python
df = pd.DataFrame(coords, columns=("name", "funicular", "lon", "lat", "network", "operator"))
```

```python
df.info()
```

```python
df['coordinates'] = "(" + df['lon'].map(str) + " , "  + df['lat'].map(str)+")" 
```
